### 1.1 Loading the relevant data in dataframes

The first step is to load the data into Pandas dataframes. I will load our database into two separate dataframes: one for the venture data and one for the cashflows. I will also load the entire TSTD in its own dataframe. Once loaded, it is easy to combine the data into new, special-purpose, dataframes for further calculations or analysis. Our two tables are linked through the 'Voyage ID' key and (where possible) also to the TSTD through the TSTD ID. This gives us the possibility to freely link the databases.

```python
# First, we import the Pandas library
import pandas as pd

# Next, we import the actual profits database. It comes in two parts: a ventures table and 
# a cashflows table
ventures = pd.read_csv('Venture database GK.csv', index_col=0)
cashflows = pd.read_csv('Cash flow database GK.csv')

# We also import the Transatlantic Slave Trade Database (TSTD, slavevoyages.org, 
# 2020 release of Voyages database, in SPSS-format)
# The TSTD source file can be downloaded from https://slavevoyages.org/voyage/downloads
# Note: in order to read SPSS-files, the Python dependency pyreadstat needs to be installed
TSTD = pd.read_spss('tastdb-exp-2020.sav')

# The TSTD has a special format for dates. This function easily converts TSTD-dates into the
# flexible Python datetime format
# Thanks to Judith for discovering the TSTD stores dates in 'seconds from 1582-10-14'
import datetime
def normalizedate(tstddate):
    """""
    Function to normalize the dates from the TSTD
    """
    return (datetime.datetime(1582, 10, 14) + datetime.timedelta(seconds = tstddate))

# The following function can be called upon to pull data from the TSTD
def pull_from_tstd(tstd_id, tstd_variable, isdate=False):
    """"
    Function to pull data from the TSTD
    tstd_id = the Voyage ID of the TSTD
    tstd_variable = the name of the variable to lookup in the TSTD (can be found in TSTD Handbook)
    isdate = optionally set to True if fetching a date, which will then be converted to Python datetime
    If function is given an empty TSTD ID or the TSTD variable holds no data, 'NA' is returned
    """
    if pd.isnull(tstd_id):
        return pd.NA
    
    if not pd.isnull(TSTD[tstd_variable].loc[(TSTD['VOYAGEID'] == tstd_id)].iloc[0]):
        if isdate == False:
            return TSTD[tstd_variable].loc[(TSTD['VOYAGEID'] == tstd_id)].iloc[0]
        else:
            return normalizedate(TSTD[tstd_variable].loc[(TSTD['VOYAGEID'] == tstd_id)].iloc[0])
    else:
        return pd.NA
```

### 1.2 Combining the databases for analysis

Now that the Dutch data has all been loaded into memory, we can start our analysis. There are many possibilities, but for the present purposes I will combine relevant data into a new, custom-built dataframe to research the ROI on slave voyages. 

Please note that my data on the Dutch ventures is not complete. In many cases, it lacks a date of return to the home port. I want to include the year in which a ship came back to see possible developments over time. Therefore, the following script pulls missing data from the TSTD. It does the same for missing data on the region of slave voyages and number of slaves acquired. This can easily be done, since all of my voyages contain TSTD IDs. The result of the following script will be a new dataframe containing all the information I need for my analysis.

```python
# Create an empty list to hold the data on the profitability of voyages. We will
# use this Python data structure to store further information on the profitability
# of individual voyages. At the end of this code cell, we convert the resulting
# list into a new dataframe
profitlist = []

# Now we will populate the list with the voyages. We first loop through
# all voyages with a unique ID in the cashflow database
for voyage in cashflows['VentureID'].unique():            
        
    # Retrieve ship name
    ship_name = ventures.loc[voyage, 'Name of the ship']
    
    # Retrieve whether data on outlays and returns are complete
    outlays_complete = ventures.loc[voyage, 'Complete data on outlays?']
    returns_complete = ventures.loc[voyage, 'Complete data on returns?']
    
    # Calculate the investment, returns and costs on return
    # Please note: in my own data, I have no information on the timing of cashflows (other than 'Outfitting' or 'Return')
    # The next version of this script will incorporate a check for timing of cashflows. If we have these, we will
    # be able to calculate IRR
    investment = cashflows['value'].loc[(cashflows['VentureID'] == voyage) & (cashflows['timing'] == 'Outfitting') & (cashflows['typeofcashflow'] == 'Expenditure')].sum()
    returns = cashflows['value'].loc[(cashflows['VentureID'] == voyage) & (cashflows['timing'] == 'Return') & (cashflows['typeofcashflow'] == 'Return')].sum()
    costonreturn = cashflows['value'].loc[(cashflows['VentureID'] == voyage) & (cashflows['timing'] == 'Return') & (cashflows['typeofcashflow'] == 'Expenditure')].sum()
 
    # Override incomplete data on outlays and returns (not useful for calculations)
    if outlays_complete != 'yes':
        investment = pd.NA
    if returns_complete != 'yes':
        returns = pd.NA
        costonreturn = pd.NA   
    
    # Use the above-mentioned variables to calculate the gross returns, result and the ROI
    gross_returns = returns - costonreturn
    result = gross_returns - investment
    ROI = result/investment * 100
    
    # Lookup the venture's TSTD ID
    tstd_id = ventures.loc[voyage, 'Voyage-ID in TSTD']
    
    # Pull the date of return from the TSTD if necessary
    if not pd.isnull(ventures.loc[voyage, 'Date of return to port of outfitting']):
        date_return = datetime.datetime.strptime(ventures.loc[voyage, 'Date of return to port of outfitting'],"%Y-%m-%d")
    else:
        date_return = pull_from_tstd(tstd_id, 'DATEEND', isdate=True)
        
    # We usually have information on the full date of return. For 
    # the following calculations, we need only the year of return
    # This is easy to extract, since all dates are now stored in
    # Python datetime format
    if not pd.isnull(date_return):
        date_return = date_return.year
    
    # Pull the (primary) region of purchase of slaves from the TSTD if necessary
    if not pd.isnull(ventures.loc[voyage, 'Place of purchase']):
        regionpurchase = ventures.loc[voyage, 'Place of purchase']
    else:
        regionpurchase = pull_from_tstd(tstd_id, 'MAJBYIMP')

    # Pull the number of slaves bought in Africa from the TSTD if necessary
    if not pd.isnull(ventures.loc[voyage, 'Number of slaves purchased']):
        number_purchased = ventures.loc[voyage, 'Number of slaves purchased']
    else:
        pull_from_tstd(tstd_id, 'SLAXIMP')
            
    # Construct a list with data on this voyage
    # It will store the following info in order: Voyage ID, ship name, year of return, 
    # region of purchase, number of slaves purchased, investment, gross returns, result, currency, ROI
    profitvoyage = [voyage, ship_name, date_return, regionpurchase, number_purchased, investment, gross_returns, result, cashflows['currency'].loc[(cashflows['VentureID'] ==  voyage)].iloc[0], ROI]
    
    # Append that list to the aforementioned list to create a list of lists
    profitlist.append(profitvoyage)
    
# Create a Pandas dataframe from the aforementioned list of lists
analysis = pd.DataFrame.from_records(profitlist, columns=['VoyageID', 'Ship', 'End of voyage', 'Region of slave purchase', 'Number of slaves', 'Investment', 'Gross returns', 'Result', 'Currency', 'ROI'], index='VoyageID')
```

We now have a dataframe called 'analysis', that contains the following information for each voyage:
1. The Voyage ID
2. The ship's name
3. The year in which the voyage ended (the underlying data often contains the day/month as well)
4. The principal region of slave purchase
5. The number of slaves purchased
6. The invested amount
7. The gross returns
8. The (financial) result
9. The currency in which the result is expressed
10. The simple ROI

If any of these datapoints are not available for a voyages, it will register as 'NA' or 'NaN'. Before our calculations, we can easily filter out the missing data if necessary. 

We can now get a first glimpse of the custom-built dataframe. If necessary, we could also export it in various formats or pretty print it by applying further styling.

### 1.3 Analyzing the ROI

The last column of the 'analysis'-dataframe contains the ROI for the voyages. Although it is not visible in the sample above, in two cases the data does not contain information on returns. In these cases, the ROI is NA. We will filter out these incomplete samples by using 'dropna()'.

We can have Pandas calculate some stats to examine the spread of the ROI-values:

```python
# Describe the data on'Result' and 'ROI'
analysis[['Result', 'ROI']].dropna().astype(int).describe().round(2).style.format(formatter={('Result'): "{:,}", ('ROI'): "{:,.2f}"})
```

We have ROI-data on 75 voyages. The mean ROI is 4.59%, but the data points are spread between -58% and +47%. Pretty volatile indeed. 

We can regroup the dataframe on any column. Let's see if there are differences in ROI for voyages to different African regions by regrouping on the principal region of slave purchase.

```python
# Regroup data on ROI by region of slave purchase. We will also add a
# count to see from how many data points each mean is calculated
check_ROI = analysis[['Region of slave purchase', 'ROI']].dropna()
check_ROI['ROI'] = check_ROI['ROI'].astype(float)
check_ROI.groupby(['Region of slave purchase']).agg(['mean', 'count']).round(2) 
```

Let's also take a closer look at the spread of the ROI data per region.

```python
# Describe the spread of the ROI data per region
check_ROI.groupby(['Region of slave purchase']).describe().round(2)
```

As the above table makes clear, in the various regions the standard deviation is rather high compared to the means. It also shows that most of the voyages went to the windward coast, followed by West Central Africa (which seems to have been substantially less profitable). 

To end this first exercise, let's make a scatter plot to view the possible development of profits over time.

```python
# Plot the ROI per year (for voyages where year data is available)
profitsintime = analysis[['End of voyage', 'ROI']].dropna().set_index('End of voyage').sort_values(by=['End of voyage']).plot(style='o')
profitsintime.set_ylabel("ROI")
profitsintime.set_xlabel("End of voyage")
```

If we want to deepen our analysis, it would of course be possible to use a function to plot a trend line. Or we could go more fancy and add colors for the various regions to the graph above.
