## 1. A first simple calculation of profitability

I will first try out the database and perform some preliminary calculations on my own data.

### 1.1 Importing the data from our database

```python
# First, we import the Pandas library to work with the databases (in CSV-format)
import pandas as pd

# Next, we import the actual profits database. It comes in two parts: a ventures table and 
# a cashflows table
ventures = pd.read_csv('Venture database GK.csv', index_col=0)
cashflows = pd.read_csv('Cash flow database GK.csv')

# We also import the Transatlantic Slave Trade Database (TSTD, slavevoyages.org, 
# 2020 release of Voyages database, in SPSS-format)
# The TSTD source file can be downloaded from https://slavevoyages.org/voyage/downloads
# Note: in order to read SPSS-files, the Python dependency pyreadstat needs to be installed
TSTD = pd.read_spss('tastdb-exp-2020.sav')

# The TSTD has a special format for dates. This function easily converts TSTD-dates into regular ones
# Thanks to Judith for discovering the TSTD stores dates in 'seconds from 1582-10-14'
import datetime
def normalizedate(tstddate):
    return (datetime.datetime(1582, 10, 14) + datetime.timedelta(seconds = tstddate))
```

The database has now been loaded in two separate Pandas 'dataframes' (comparable to Excel tables, but more powerful). The entire TSTD Voyages database has also been loaded in a dataframe. Our two tables are linked through the 'Voyage ID' key and (where possible) also to the TSTD through the TSTD ID. This gives us the possibility to freely link the databases and create new dataviews for calculation and visualization purposes.

### 1.2 Calculate the ROI for each voyage

```python
# Create an empty list to hold the data on the profitability of voyages. We will
# use this Python data structure to store further information on the profitability
# of individual voyages. At the end of this code cell, we convert the resulting
# list into a new dataframe
profitlist = []

# Now we will populate the list with the voyages. We first loop through
# all voyages with a unique ID in the cashflow database
for voyage in cashflows['VentureID'].unique():
   
    # We first check to see if we have full information on the voyage
    # If not, we move on to the next voyage
    # Note that the Python keyword 'continue' here means: continue to the next voyage
    if (ventures.loc[voyage, 'Complete data on outlays?'] == 'yes' and ventures.loc[voyage, 'Complete data on returns?'] == 'yes') == False:
        continue
                                          
    # Calculate the investment, returns and costs on return
    # Please note: in my own data, I have no information on the timing of cashflows (other than 'Outfitting' or 'Return')
    # The next version of this script will incorporate a check for timing of cashflows. If we have these, we will
    # be able to calculate IRR
    investment = cashflows['value'].loc[(cashflows['VentureID'] == voyage) & (cashflows['timing'] == 'Outfitting') & (cashflows['typeofcashflow'] == 'Expenditure')].sum()
    returns = cashflows['value'].loc[(cashflows['VentureID'] == voyage) & (cashflows['timing'] == 'Return') & (cashflows['typeofcashflow'] == 'Return')].sum()
    costonreturn = cashflows['value'].loc[(cashflows['VentureID'] == voyage) & (cashflows['timing'] == 'Return') & (cashflows['typeofcashflow'] == 'Expenditure')].sum()

    # Use the above-mentioned variables to calculate the result and the ROI
    result = returns - costonreturn - investment
    ROI = result/investment * 100
    
    # Lookup the venture's TSTD ID
    tstdid = ventures.loc[voyage, 'Voyage-ID in TSTD']
    
    # Check if our database holds a date of return (end of voyage)
    # If not: see if the TSTD has any date and use that one
    # If both our database and the TSTD doesn't have a date for the end of the voyage, simply register it as 'NA'
    if not pd.isnull(ventures.loc[voyage, 'Date of return to port of outfitting']):
        dateofreturn = datetime.datetime.strptime(ventures.loc[voyage, 'Date of return to port of outfitting'],"%Y-%m-%d").year
    elif not pd.isnull(TSTD['DATEEND'].loc[(TSTD['VOYAGEID'] == tstdid)].iloc[0]):
        dateofreturn = normalizedate(TSTD['DATEEND'].loc[(TSTD['VOYAGEID'] == tstdid)].iloc[0]).year
    else:
        dateofreturn = pd.NA
    
    # Use the same procedure for the (primary) region of purchase of slaves
    if not pd.isnull(ventures.loc[voyage, 'Place of purchase']):
        placepurchase = ventures.loc[voyage, 'Place of purchase']
    elif not pd.isnull(TSTD['MAJBYIMP'].loc[(TSTD['VOYAGEID'] == tstdid)].iloc[0]):
        placepurchase = TSTD['MAJBYIMP'].loc[(TSTD['VOYAGEID'] == tstdid)].iloc[0]
    else:
        placepurchase = pd.NA
            
    # Construct a list with data on this voyage
    # It will store the following info in order: Voyage ID, TSTD ID, Ship name, Region of purchase, year of return, result, currency, ROI
    profitvoyage = [voyage, tstdid, ventures.loc[voyage, 'Name of the ship'], placepurchase, dateofreturn, investment, returns - costonreturn, result, cashflows['currency'].loc[(cashflows['VentureID'] ==  voyage)].iloc[0], ROI]
    
    # Append that list to the aforementioned list to create a list of lists
    profitlist.append(profitvoyage)
    
# Create a Pandas dataframe from the aforementioned list of lists
profitability = pd.DataFrame.from_records(profitlist, columns=['VoyageID', 'TSTD', 'Ship', 'Region of slave purchase', 'End of voyage', 'Investment', 'Gross returns', 'Result', 'Currency', 'ROI'], index='VoyageID')
```

We now have a dataframe called 'profitability', that contains the following information for each voyage:
1. The Voyage ID
2. The TSTD ID
3. The ship's name
4. The principal region of slave purchase
5. The year in which the voyage ended (the underlying data often contains the day/month as well)
6. The (financial) result
7. The currency in which the result is expressed
8. The simple ROI

The principal region of slave purchase and the year in which the voyage ended are both taken from the TSTD, if our database does not contain these data for individual voyages.

### 1.3 First analysis

We can now get a first glimpse of this dataframe. If necessary, we could export the following table in various formats. We could also pretty print it by applying further styling.


```python
# Display the dataframe
profitability
```

We can have Pandas calculate some data parameters to examine the spread of the values:

```python
# Describe the data on'Result' and 'ROI'
profitability[['Result', 'ROI']].describe().style.format(formatter={('Result'): "{:,}", ('ROI'): "{:,.2f}"})
```

We can also regroup the dataframe on any column. For example, let's regroup on the principal region of slave purchase, to see if there are major differences between regions in terms of ROI. 

```python
# Regroup data on ROI by region of slave purchase. We will also add a
# count to see from how many data points each mean is calculated
profitability[['Region of slave purchase', 'ROI']].groupby(['Region of slave purchase']).agg(['mean', 'count']).round(2) 
```

Let's also take a closer look at the spread of the ROI data per region.

```python
# Describe the spread of the ROI data per region
profitability[['Region of slave purchase', 'ROI']].groupby(['Region of slave purchase']).describe().round(2)
```

As the above table makes clear, in the various regions the standard deviation is rather high compared to the means. The results of the individual voyages were probably highly volatile.

To end this first exercise, let's make a scatter plot to view the possible development of profits over time.

```python
# Plot the ROI per year (for voyages where year data is available)
profitsintime = profitability[['End of voyage', 'ROI']].dropna().set_index('End of voyage').sort_values(by=['End of voyage']).plot(style='o')
profitsintime.set_ylabel("ROI")
profitsintime.set_xlabel("End of voyage")
```
